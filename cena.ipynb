```python
from datetime import datetime

from pymongo import MongoClient, DESCENDING

server = "192.168.2.68"
port = '27018'
TOTALPOWER = ["TotalPower","powerrightside"]
TOTALPOWERHOUR = ["BuildingLeftSide","totalpowerhour"]
start = datetime.strptime("2024-03-27 00:00:00", "%Y-%m-%d %H:%M:%S")
end = datetime.strptime("2024-03-28 00:00:00", "%Y-%m-%d %H:%M:%S")

client = MongoClient(server + ':' + port)
building_totalpower = list(
client[TOTALPOWERHOUR[0]][TOTALPOWERHOUR[1]].find({'datetime': {'$gt': start, '$lt': end}}))
client.close()
building_totalpower
```

```python
consumption = []
for line in building_totalpower:
    consumption.append([line['datetime'],line['consumption']])
```

```python
import pandas as pd
total = pd.DataFrame(consumption, columns=['datetime', 'consumption'])
total['datetime'] = pd.to_datetime(total['datetime'], format='%Y-%m-%d %H:%M:%S', dayfirst=True)
total.set_index("datetime", inplace=True)
total = total.resample('1h').mean()
total
```
